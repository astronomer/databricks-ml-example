```python
target_variable = dbutils.widgets.get("target_variable")
features_db = dbutils.widgets.get("features_database")
features_table = dbutils.widgets.get("features_table")
data_dir = dbutils.widgets.get("data_dir")
```

```python
import databricks.automl
import mlflow
import pandas
import sklearn.metrics
```

```python
input_df = spark.sql(f"SELECT * FROM {features_db}.{features_table}")
```

```python
train_df, test_df = input_df.randomSplit([0.99, 0.01], seed=55)
display(train_df)
```

```python
summary = databricks.automl.classify(train_df, target_col=target_variable, data_dir=data_dir, timeout_minutes=30)
```
