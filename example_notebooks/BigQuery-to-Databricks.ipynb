```python
spark.conf.set("viewsEnabled","true")

bq_project_id = "<bq_project_id>"


sql = """
  SELECT
    GENERATE_UUID() uuid,
    age,
    REPLACE(TRIM(workclass), '?', 'Unknown') AS workclass,
    REPLACE(TRIM(education), '?', 'Unknown') AS education,
    REPLACE(TRIM(marital_status), '?', 'Unknown') AS marital_status,
    REPLACE(TRIM(occupation), '?', 'Unknown') AS occupation,
    REPLACE(TRIM(race), '?', 'Unknown') AS race,
    REPLACE(TRIM(sex), '?', 'Unknown') AS sex,
    capital_gain,
    capital_loss,
    hours_per_week,
    REPLACE(TRIM(native_country), '?', 'Unknown') AS native_country,
    REPLACE(TRIM(income_bracket), '?', 'Unknown') AS income_bracket
  FROM
    `bigquery-public-data.ml_datasets.census_adult_income`
  WHERE
    age IS NOT NULL
    AND workclass IS NOT NULL
    AND education IS NOT NULL
    AND marital_status IS NOT NULL
    AND occupation IS NOT NULL
    AND race IS NOT NULL
    AND sex IS NOT NULL
    AND capital_gain IS NOT NULL
    AND capital_loss IS NOT NULL
    AND hours_per_week IS NOT NULL 
    AND native_country IS NOT NULL
    AND income_bracket IS NOT NULL
"""


df = spark.read.format("bigquery") \
    .option("materializationDataset","gcs_to_bq_operator") \
    .option("project", bq_project_id) \
    .option("parentProject", bq_project_id) \
    .option("query", sql) \
    .load()

df = df.drop_duplicates()
```

```python
df.write.format("parquet").saveAsTable("census_adult_income")
```
